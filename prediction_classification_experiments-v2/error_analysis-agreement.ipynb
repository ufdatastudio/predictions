## Agreement vs Accuracy

```python
df.sort_values('Per Row Agreement')
```

```python
df.loc[:, ['Base Sentence', 'Sentence Label', 'Per Row Agreement']].sort_values('Per Row Agreement')
```

```python
df['Majority Vote'] = df[model_col_names].mode(axis=1)
df
```

```python
def classification_of_mv_sentence(df):
    copy_df = df.copy()
    copy_df['Class'] = None

    for idx, row in copy_df.iterrows():
        y = row['Sentence Label']
        y_hat = row['Majority Vote']

        # y = 1 (positive class)
        if y == 1:
            if y_hat == 1:  # TP
                copy_df.loc[idx, 'Class'] = "TP"
            else:  # FN
                copy_df.loc[idx, 'Class'] = "FN"

        # y = 0 (negative class)
        elif y == 0:
            if y_hat == 0:  # TN
                copy_df.loc[idx, 'Class'] = "TN"
            else:  # FP
                copy_df.loc[idx, 'Class'] = "FP"

    return copy_df
```

```python
mv_with_class_df = classification_of_mv_sentence(df)
mv_with_class_df
```

```python
filt_tp = (mv_with_class_df.sort_values('Per Row Agreement')['Class'] == "TP")
tp_df = mv_with_class_df.sort_values('Per Row Agreement')[filt_tp]
tp_df
```

```python
filt_fn = (mv_with_class_df.sort_values('Per Row Agreement')['Class'] == "FN")
fn_df = mv_with_class_df.sort_values('Per Row Agreement')[filt_fn]
fn_df
```

```python
# Total prediction class (TP + FN)
prediction_class_total = len(tp_df) + len(fn_df)

# TP percentage
tp_percentage = round(len(tp_df) / prediction_class_total * 100, 2)

# FN percentage
fn_percentage = round(len(fn_df) / prediction_class_total * 100, 2)

print(f"Recall for Prediction Class Total (TP + FN): {prediction_class_total}")
print(f"TP: {len(tp_df)} ({tp_percentage}%)")
print(f"FN: {len(fn_df)} ({fn_percentage}%)")
```
