```python
import os
import sys

# Get the current working directory of the notebook
notebook_dir = os.getcwd()
# print(notebook_dir)
# Add the parent directory to the system path
sys.path.append(os.path.join(notebook_dir, '../'))

from metrics import EvaluationMetric
from data_processing import DataProcessing
from feature_extraction import SpacyFeatureExtraction
```

```python
base_data_path = DataProcessing.load_base_data_path(notebook_dir)
data_path = os.path.join(base_data_path, 'extract_prediction_properties/future-chronicle2050_fpb_synthethic-v1/openai/gpt-oss-120b/results.csv')
df = DataProcessing.load_from_file(data_path)
df
```

```python
y_df = df.copy()
y_hat_df = df.copy()
```

```python
y_spacy_fe = SpacyFeatureExtraction(y_df, 'Source')
embed_source_y_df = y_spacy_fe.sentence_embeddings_extraction(attach_to_df=True)

y_hat_spacy_fe = SpacyFeatureExtraction(y_hat_df, 'Source')
embed_source_y_hat_df = y_hat_spacy_fe.sentence_embeddings_extraction(attach_to_df=True)
```

```python
import numpy as np
np.testing.assert_equal(len(embed_source_y_df), len(embed_source_y_hat_df))
```

```python
len_dfs = len(embed_source_y_df)

#          y, y_hat
tps = [] # 1, 1 =>
fns = [] # 1, 0 => 
tns = [] # 0, 0 =>
fps = [] # 0, 1 =>
misc = []

for idx in range(len_dfs):
    # print(idx)
    y_embed = embed_source_y_df['Source Embedding'][idx]
    y_hat_embed = embed_source_y_hat_df['Source Embedding'][idx]

    if idx < 3:
        print(f"{type(y_embed)} vs {type(y_hat_embed)}")
    
    if y_embed is None and y_hat_embed is None:
        tns.append(idx)

    if y_embed is not None and y_hat_embed is None:
        fns.append(idx)

    if y_embed is None and y_hat_embed is not None:
        fps.append(idx)

    if y_embed is not None and y_hat_embed is not None:
        x_embeddings_reshaped = y_embed[idx].reshape(1, -1)
        y_embeddings_reshaped = y_hat_embed[idx].reshape(1, -1)
        cs = EvaluationMetric.get_cosine_similarity(x_embeddings_reshaped, y_embeddings_reshaped, per_row=False)
        if idx < 3:
            # print(f"{y_embed} vs {y_hat_embed} => {cs}")
            print(f"CS: {cs}")
            if cs == 1:
                tps.append(idx)
            else:
                misc.append(idx)
```
