```python
import warnings
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
warnings.filterwarnings("ignore")
```

```python
X, y = make_classification(
    n_samples=1000,       # Total number of samples
    n_features=20,        # Number of features
    n_informative=2,      # Number of informative features
    n_redundant=10,       # Number of redundant features
    n_clusters_per_class=1,
    weights=[0.905, 0.095],   # Imbalance ratio
    flip_y=0,             # No noise
    random_state=1        # Reproducibility
)
X
```

```python
y
```

```python
import pandas as pd
y_df = pd.DataFrame(y)
y_df.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train
```

```python
X_test
```

```python
class_weights = []
# custom_class_weight_1 = {A: 10, B: 1}
custom_class_weight_2 = {A: 1, B: 10}
custom_class_weight_3 = {A: 1.5, B: .75}


class_weights.append('balanced')
class_weights.append(custom_class_weight_1)
class_weights.append(custom_class_weight_2)
class_weights.append(custom_class_weight_3)
class_weights
```

```python
for class_weight in class_weights:
    # print("="*70)
    print(f"\t\tclass_weight: {class_weight}")
    model = LogisticRegression(class_weight=class_weight, max_iter=1000, solver='lbfgs')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    # Evaluate the model
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("="*70)
```
