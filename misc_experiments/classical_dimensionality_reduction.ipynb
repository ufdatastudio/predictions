```python
import numpy as np
from sklearn.decomposition import PCA
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
X
```

```python
pca = PCA(n_components=2)
pca
```

```python
pca.fit(X)
```

```python
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```
